```python
with open('cnn_1min.txt', 'rb') as f:
    response = pickle.load(f)
    
response
```

```python
# response.results[0].alternatives[0].words[0].word

for result in response.results[:4]:
    if result.channel_tag == 1:
        print('\n')
        for alternative in result.alternatives:
            for word in alternative.words:
                print(word.word, end=' ')
```

```python
for result in response.results[:4]:
    if result.channel_tag == 1:
        print('\n')
        for alternative in result.alternatives:
            for word in alternative.words:
                start_time = word.start_time.seconds + word.start_time.nanos / 10**9
                end_time = word.end_time.seconds + word.end_time.nanos / 10**9
                print(f'{start_time}~{end_time} : {word.word}')
```
